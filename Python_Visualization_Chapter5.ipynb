## Exercise 1: Sales by Region (Bar Chart)

```python
import matplotlib.pyplot as plt

regions = ['North','South','East','West','Central']
sales = [25000,18000,22000,27000,20000]

plt.figure(figsize=(8,5))
plt.bar(regions, sales)
plt.title('Company Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales (USD)')
plt.grid(axis='y')
plt.show()
```

## Exercise 2: Monthly Temperature Trend (Line Chart)

```python
months = ['Jan','Feb','Mar','Apr','May','Jun']
temperature = [24,26,28,30,33,35]

plt.figure(figsize=(8,5))
plt.plot(months, temperature, marker='o')
plt.title('Monthly Temperature Trend')
plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.show()
```
