# 1.1 Introduction to Data Visualization

Data visualization is the graphical representation of data using charts, graphs, and other visual elements.
It helps users understand trends, patterns, and insights quickly.


````python

import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr"]
sales = [10000, 12000, 15000, 18000]

plt.plot(months, sales, marker='o')
plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
````


# 1.2 Importance of Visual Storytelling


# 1.3 Types of Charts and When to Use Them



## Line Chart
Used to show trends over time.


````python

import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May"]
sales = [10, 15, 20, 25, 30]

plt.plot(months, sales, marker='o')
plt.title("Monthly Sales Trend")
plt.show()
````


## Bar Chart
Used to compare categories.


````python

import matplotlib.pyplot as plt

products = ["A", "B", "C", "D"]
sales = [50, 80, 65, 95]

plt.bar(products, sales)
plt.title("Product Sales")
plt.show()
````


## Pie Chart
Used to show percentage contribution.


````python

import matplotlib.pyplot as plt

labels = ["Marketing", "HR", "IT", "Operations"]
sizes = [30, 15, 25, 30]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Department Budget Allocation")
plt.show()
````


## Histogram
Used to show frequency distribution.


````python

import matplotlib.pyplot as plt

scores = [55,60,62,65,67,70,72,75,80,85,90,95]

plt.hist(scores, bins=5)
plt.title("Student Score Distribution")
plt.show()
````


## Scatter Plot
Used to analyze relationships between variables.


````python

import matplotlib.pyplot as plt

study_hours = [1,2,3,4,5,6]
marks = [40,50,55,65,75,90]

plt.scatter(study_hours, marks)
plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
````

# Introduction

Data visualization helps transform raw data into meaningful insights.

# 3.1 Line Charts

Line charts are used to show trends and changes over time.

### Example: Monthly Sales Trend


````python

import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [10000, 12000, 15000, 18000, 21000, 25000]

plt.plot(months, sales, marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
````


### Interpretation
- Sales increase steadily each month.
- June shows the highest sales.



# 3.2 Bar Charts

Bar charts compare values across categories.

### Example: Product Sales Comparison


````python

import matplotlib.pyplot as plt

products = ["Laptop", "Mobile", "Tablet", "Monitor"]
sales = [120, 200, 90, 150]

plt.bar(products, sales)
plt.title("Product Sales")
plt.xlabel("Products")
plt.ylabel("Units Sold")
plt.show()
````


### Interpretation
- Mobile has the highest sales.
- Tablet has the lowest sales.



# 3.3 Pie Charts

Pie charts show proportions of a whole.

### Example: Department Budget Allocation


````python

import matplotlib.pyplot as plt

departments = ["HR", "IT", "Marketing", "Finance"]
budget = [10, 35, 30, 25]

plt.pie(budget, labels=departments, autopct='%1.1f%%')
plt.title("Department Budget Allocation")
plt.show()
````


### Interpretation
- IT receives the highest budget allocation.
- HR receives the smallest share.



# 3.4 Histograms

Histograms show the distribution of numerical data.

### Example: Student Score Distribution


````python

import matplotlib.pyplot as plt

scores = [55,60,62,65,68,70,72,75,78,80,82,85,88,90,92,95]

plt.hist(scores, bins=5)
plt.title("Student Score Distribution")
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.show()
````


### Interpretation
- Most students score between 70 and 90.



# 3.5 Scatter Plots

Scatter plots identify relationships between variables.

### Example: Study Hours vs Marks


````python

import matplotlib.pyplot as plt

study_hours = [1,2,3,4,5,6,7,8]
marks = [35,42,50,60,68,75,85,92]

plt.scatter(study_hours, marks)
plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
````


### Interpretation
- Positive correlation exists between study hours and marks.



# Comparison of Chart Types

## Exercise 2: Add Markers and Gridlines

````python
plt.plot(months, sales, marker='o')
plt.title('Monthly Sales Performance')
plt.xlabel('Months')
plt.ylabel('Sales (USD)')
plt.grid(True)
plt.show()
````

## Exercise 3: Professional Bar Chart

````python
products = ['Laptop','Mobile','Tablet','Monitor']
units = [120,200,90,150]

plt.figure(figsize=(8,5))
plt.bar(products, units)
plt.title('Product Sales Analysis')
plt.xlabel('Products')
plt.ylabel('Units Sold')
plt.grid(axis='y')
plt.show()
````

## Exercise 4: Multiple Series with Legend

````python
sales_2025 = [12,15,13,17,20,23]
sales_2026 = [14,18,16,21,24,28]

plt.plot(months, sales_2025, marker='o', label='2025')
plt.plot(months, sales_2026, marker='s', label='2026')
plt.title('Sales Comparison')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
````
