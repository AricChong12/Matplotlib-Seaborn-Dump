```python
import matplotlib.pyplot as plt

months = ['Jan','Feb','Mar','Apr','May','Jun']
sales = [12000,15000,13000,17000,20000,23000]

plt.plot(months, sales)
plt.show()
```

## Exercise 1: Add Titles and Labels

```python
plt.plot(months, sales)
plt.title('Monthly Sales Performance')
plt.xlabel('Months')
plt.ylabel('Sales (USD)')
plt.show()
```

## Exercise 2: Add Markers and Gridlines

```python
plt.plot(months, sales, marker='o')
plt.title('Monthly Sales Performance')
plt.xlabel('Months')
plt.ylabel('Sales (USD)')
plt.grid(True)
plt.show()
```

## Exercise 3: Professional Bar Chart

```python
products = ['Laptop','Mobile','Tablet','Monitor']
units = [120,200,90,150]

plt.figure(figsize=(8,5))
plt.bar(products, units)
plt.title('Product Sales Analysis')
plt.xlabel('Products')
plt.ylabel('Units Sold')
plt.grid(axis='y')
plt.show()
```

## Exercise 4: Multiple Series with Legend

```python
sales_2025 = [12,15,13,17,20,23]
sales_2026 = [14,18,16,21,24,28]

plt.plot(months, sales_2025, marker='o', label='2025')
plt.plot(months, sales_2026, marker='s', label='2026')
plt.title('Sales Comparison')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```
