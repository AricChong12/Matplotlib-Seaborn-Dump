## Activity 1: Sales Performance Analysis

**Explanation:** Loads a CSV file and creates a bar chart of monthly sales.

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('sales_data.csv')
print(df)

plt.bar(df['Month'],df['Sales'])

plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales (RM)')
plt.show()
```

## Activity 2: Student Marks Analysis

**Explanation:** Loads an Excel file and visualizes student marks.

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_excel('student_marks.xlsx')
print(df)
plt.bar(df['Student'],df['Marks'])
plt.title('Student Marks')
plt.show()
```

## Activity 3: Product Category Sales

**Explanation:** Creates a pie chart showing sales contribution by category.

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('product_sales.csv')
plt.pie(df['Sales'],labels=df['Category'],autopct='%1.1f%%')
plt.title('Sales by Product Category')
plt.show()
```

## Activity 4: Employee Salary Distribution

**Explanation:** Creates a histogram to show salary distribution.

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('employee_salary.csv')
plt.hist(df['Salary'],bins=5)
plt.title('Employee Salary Distribution')
plt.show()
```

## Activity 5: Website Traffic Trend

**Explanation:** Creates a line chart of daily website visitors.

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_excel('website_visitors.xlsx')
plt.plot(df['Day'],df['Visitors'],marker='o')
plt.grid(True)
plt.title('Website Visitors')
plt.show()
```

## Activity 6: Future Sales Prediction

### Objective
Predict the next value using historical data with **Linear Regression**.

### How it works
1. Load the dataset using Pandas.
2. Create X values (1,2,3...) representing the order of records.
3. Use **Sales** as the target variable (y).
4. Train a Linear Regression model.
5. Predict the next value.
6. Plot the actual data, trend line, and predicted value.

### Expected Learning
- Read CSV/Excel files.
- Train a simple machine learning model.
- Make a future prediction.
- Interpret the visualization.


```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression

# Step 1: Load dataset
df = pd.read_csv("sales_data.csv")
print(df)

# Step 2: Create input (X) and output (y)
X = [[i] for i in range(1, len(df)+1)]
y = df["Sales"]

# Step 3: Train the model
model = LinearRegression()
model.fit(X, y)

# Step 4: Predict the next value
prediction = model.predict([[len(df)+1]])
print("Predicted Sales:", round(prediction[0],2))

# Step 5: Visualize results
plt.scatter(X, y, label="Actual Data")
plt.plot(X, model.predict(X), label="Trend Line")
plt.scatter(len(df)+1, prediction, color="red", s=100, label="Prediction")
plt.title("Future Sales Prediction")
plt.xlabel("Observation")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.show()
```
