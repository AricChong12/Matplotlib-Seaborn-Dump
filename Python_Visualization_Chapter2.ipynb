# 2.2 Basic Setup and Plotting Simple Graphs



## Example 1: Matplotlib Line Chart


````

import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May"]
sales = [10, 15, 18, 25, 30]

plt.plot(months, sales, marker='o')
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
````


## Example 2: Matplotlib Bar Chart


````

import matplotlib.pyplot as plt

products = ["Laptop", "Mobile", "Tablet", "Monitor"]
sales = [120, 200, 90, 150]

plt.bar(products, sales)
plt.title("Product Sales")
plt.xlabel("Products")
plt.ylabel("Units Sold")
plt.show()
````


## Example 3: Pie Chart


````

import matplotlib.pyplot as plt

departments = ["HR", "IT", "Marketing", "Finance"]
budget = [10, 35, 30, 25]

plt.pie(budget, labels=departments, autopct='%1.1f%%')
plt.title("Department Budget Allocation")
plt.show()
````


## Example 4: Histogram


````

import matplotlib.pyplot as plt

scores = [55,60,65,68,70,72,75,78,80,85,88,90,92,95]

plt.hist(scores, bins=5)
plt.title("Score Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
````


## Example 5: Scatter Plot


````

import matplotlib.pyplot as plt

study_hours = [1,2,3,4,5,6,7]
marks = [35,45,50,60,70,80,90]

plt.scatter(study_hours, marks)
plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
````


## Example 6: Seaborn Line Plot


````

import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme()

months = ["Jan","Feb","Mar","Apr","May"]
sales = [10,15,18,25,30]

sns.lineplot(x=months, y=sales, marker='o')
plt.title("Monthly Sales using Seaborn")
plt.show()
````


## Example 7: Seaborn Bar Plot


````

import seaborn as sns
import matplotlib.pyplot as plt

products = ["Laptop","Mobile","Tablet","Monitor"]
sales = [120,200,90,150]

sns.barplot(x=products, y=sales)
plt.title("Product Sales using Seaborn")
plt.show()
````
